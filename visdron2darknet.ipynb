```python
def anno2lab(txt_file_name):
    buff=[]
    with open('annotations/'+txt_file_name,'r+') as txtfile:

        for line in txtfile.readlines():
            line=line.split(',')
            #print(line)
            if line[0]=='0':
                line[0]='1'
            if line[1]=='0':
                line[1]='1'
            if line[5]=='1' or line[5]=='2':
                line[5]='1'
                line=line[5:6]+line[0:4]
                buff.append(' '.join(line))           
            elif line[5]=='4' or line[5]=='5' or line[5]=='6' or line[5]=='9':
                line[5]='0'
                line=line[5:6]+line[0:4]
                buff.append(' '.join(line))
            else:
                continue
    with open('labels/'+txt_file_name,'w') as newtxtfile:
        for line in buff:
            newtxtfile.write('%s\n'%line)
        
    
```

```python
for txtfile_name in os.listdir('annotations'):
    anno2lab(txtfile_name)
```

```python
annos=os.listdir('labels')
for ann in annos:
    temp=ann.split('_')
    #print(temp)
    if temp[0]=='0' or temp[0]=='1' or temp[0]=='2' or temp[0]=='3':
        continue
    #open the image correspond to label
    image_name='images/'+ann[:-4]+'.png'
    img=cv2.imread(image_name)
    h, w, _= img.shape
    dw=1.0/w
    dh=1.0/h
    #print(image_name,height,width)
    buff=[]
    #open correspond label
    print('open file: ','labels/'+ann)
    with open('labels/'+ann,'r+') as textfile:
        #modify each line in label
        for line in textfile.readlines():
            templine=[0]*5
            line=line.split(' ')
            templine[0]=line[0]
            templine[1]=str((float(line[1])+float(line[3]))/2*dw)
            templine[2]=str((float(line[2])+float(line[4]))/2*dh)
            templine[3]=str(float(line[3])*dw)
            templine[4]=str(float(line[4][:-1])*dh)
            #print(line[4][:-1])
            t=' '.join(templine)
            buff.append(t)
    #write to result
    print('write file: ','new_labels/'+ann)
    with open('new_labels/'+ann,'w') as newtxtfile:
        for l in buff:
            newtxtfile.write('%s\n'%l)    
    
```
